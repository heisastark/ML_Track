```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

mushroom = pd.read_csv('mushrooms.csv')

mushroom
```

```python
mushroom2 = pd.get_dummies(mushroom)
mushroom2
```

```python
X = mushroom2.iloc[:, 2:]
y = mushroom2.iloc[:,1]

print(X)
```

```python
print(y)
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, random_state=3)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
dt.score(X_train, y_train)
```

```python
dt.score(X_test, y_test)
```
